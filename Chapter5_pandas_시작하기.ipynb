**5.1.2 DataFrame**

```python
data = {'state':['Ohio','Ohio','Ohio','Nevada','Nevada','Nevada'],
        'year':[2000,2001,2002,2001,2002,2003],
        'pop':[1.5, 1.7, 3.6, 2.4, 2.9, 3.2]}
frame = pd.DataFrame(data)

frame
```

```python
frame.head()
```

```python
pd.DataFrame(data, columns=['year', 'state', 'pop'])
```

```python
frame2 = pd.DataFrame(data, columns=['year', 'state', 'pop', 'debt'],
                      index=['one','two','three','four','five','six'])
frame2
```

```python
frame2.columns
```

DataFrame의 컬럼은 Series처럼 사전 형식의 표기법으로 접근하거나 속성 형식으로 접근할 수 있다.

```python
frame2['state']
```

```python
frame2.year
```

행(로우)은 위치나 loc속성을 이용해서 이름을 통해 접근할 수 있다.

```python
frame2.loc['three']
```

```python
frame2['debt'] = 16.5
frame2
```

```python
frame2['debt'] = np.arange(6.)      #리스트나 배열을 컬럼에 대입할 때는 대입하려는 값의 길이가 DataFrame의 크기와 동일해야한다.
frame2
```

```python
val = pd.Series([-1.2, -1.5, -1.7], index=['two','four','five'])    #Series를 대입하면 DataFrame의 색인에 따라 값이 대입되며 존재하지 않는 색인에는 결측치가 대입된다.
frame2['debt'] = val
frame2
```

```python
frame2['eastern'] = frame2.state == 'Ohio'      #새로운 컬럼은 frame2.eastern 형태의 문법으로는 생성되지 않는다.
frame2
```

**5.2 핵심 기능**

**5.2.1 재색인**

```python
obj = pd.Series([4.5, 7.2, -5.3, 3.6], index=['d','b','a','c'])
obj
```

```python
obj2 = obj.reindex(['a','b','c','d','e'])
obj2
```

```python
obj3 = pd.Series(['blue','purple','yellow'], index=[0,2,4])
obj3
```

```python
obj3.reindex(range(6), method='ffill')    #reindex하며 누락된 값을 직전의 값으로 채워넣을 수 있다.
```
