**5.2.1 재색인**

```python
import numpy as np
import pandas as pd
```

```python
obj = pd.Series([4.5, 7.2, -5.3, 3.6], index = ['d','b','a','c'])
obj
```

```python
obj2 = obj.reindex(['a','b','c','d','e'])
obj2
```

```python
obj3 = pd.Series(['blue','purple','yellow'],index=[0,2,4])
obj3
```

```python
obj3.reindex(range(6), method='ffill')      #method='ffill': reindex하며 누락된 값을 직전의 값으로 채워 넣는다.
obj3.reindex(range(6), method='bfill')      #method='bfill': reindex하며 누락된 값을 다음의 값으로 채워 넣는다.
```

```python
frame = pd.DataFrame(np.arange(9).reshape(3,3),
                     index=['a','c','d'],
                     columns=['Ohio', 'Texas', 'California'])
frame
```

```python
frame2 = frame.reindex(['a','b','c','d'])
frame2
```

```python
states = ['Texas', 'Utah', 'California']
frame.reindex(columns=states)
```

**5.2.2 하나의 로우나 컬럼 삭제하기**

```python
obj = pd.Series(np.arange(5.), index=['a','b','c','d','e'])
obj
```

```python
new_obj = obj.drop('c')
new_obj
```

```python
obj.drop(['d','c'])
```

```python
data = pd.DataFrame(np.arange(16).reshape((4,4)),
                    index=['Ohio','Colorado','Utah','New York'],
                    columns=['one','two','three','four'])
data
```

```python
data.drop(['Colorado','Ohio'])      #axis=0 or axis='rows'가 생략된 것이다.
```

```python
data.drop(['two'],axis=1)           #axis='columns'로도 표현 가능하다.
```

```python
data.drop(['two','four'],axis='columns')
```

drop함수처럼 Series나 DataFrame의 크기 또는 형태를 변경하는 함수는 새로운 객체를 반환하는 대신 원본 객체를 변경한다.

```python
obj.drop('c',inplace=True)
obj
```

inplace옵션을 사용하는 경우 버려지는 값을 모두 삭제하므로 주의해서 사용해야한다.

**5.2.5 산술 연산과 데이터 정렬**

다른 index를 가지고 있는 객체 간의 산술연산이 가능하다.

```python
s1 = pd.Series([7.3, -2.5, 3.4, 1.5], index = ['a','c','d','e'])
s2 = pd.Series([-2.1, 3.6, -1.5, 4, 3.1], index = ['a','c','e','f','g'])
```

```python
s1
```

```python
s2
```

```python
s1+s2
```

```python
df1 = pd.DataFrame(np.arange(9.).reshape((3,3)), columns=list('bcd'),index=['Ohio','Texas','Colorado'])
df2 = pd.DataFrame(np.arange(12.).reshape((4,3)), columns=list('bde'),index=['Utah','Ohio','Texas','Oregon'])
```

```python
df1
```

```python
df2
```

```python
df1+df2       #양쪽에 다 존재하지 않는 라벨에 대해서는 없는 값으로 나타난다.
```

```python
df1 = pd.DataFrame({'A':[1,2]})
df2 = pd.DataFrame({'B':[3,4]})
```

```python
df1
```

```python
df2
```

```python
df1-df2       #공통되는 부분만 원소별로 계산한다.
```

**산술 연산 메서드에 채워 넣을 값 지정하기**

```python
df1 = pd.DataFrame(np.arange(12.).reshape((3,4)), columns=list('abcd'))
df2 = pd.DataFrame(np.arange(20.).reshape(4,5), columns=list('abcde'))
df2.loc[1,'b'] = np.nan
```

```python
df1
```

```python
df2
```

```python
df1 + df2
```

```python
df1.add(df2,fill_value=0)        #fill_value=0의 의미: 겹치지 않는 부분은 0으로 하여 덧셈을 한다.
```

```python
1/df1
```

```python
df1.rdiv(5)       #5/df1
```

```python
df1.div(5)        #df1/5
```

```python
df1.reindex(columns=df2.columns)
```

```python
df1.reindex(columns=df2.columns, fill_value=0)
```
