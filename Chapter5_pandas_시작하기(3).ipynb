**5.2.6 함수 적용과 매핑**

```python
import numpy as np
import pandas as pd
```

```python
frame = pd.DataFrame(np.random.randn(4,3), columns=list('bde'), index=['Utah','Ohio','Texas','Oregon'])
frame
```

```python
np.abs(frame)
```

```python
f = lambda x: x.max() - x.min()     #lambda: 한줄로 함수 정의하는 법
```

```python
frame.apply(f)
```

```python
frame.apply(f, axis=1)
```

```python
def f(x):
  return pd.Series([x.min(),x.max()], index=['min','max'])

frame.apply(f)
```

```python
format = lambda x: '%.2f' %x
frame.applymap(format)          #각 원소에 적용할 함수를 지정하기 위한 map
```

```python
frame['e'].map(format)
```

**5.2.7 정렬과 순위**

```python
obj = pd.Series(range(4), index=['d','a','b','c'])
obj.sort_index()
```

```python
frame = pd.DataFrame(np.arange(8).reshape((2,4)), index=['three', 'one'], columns=['d','a','b','c'])
frame
```

```python
frame.sort_index()
```

```python
frame.sort_index(axis=1)
```

```python
frame.sort_index(axis=1, ascending=False)     # 기본: 오름차순
```

```python
obj = pd.Series([4,7,-3,2])
obj.sort_values()
```

```python
obj = pd.Series([4, np.nan, 7, np.nan, -3, 2])
obj.sort_values()
```

```python
frame = pd.DataFrame({'b':[4,7,-3,2], 'a':[0,1,0,1]})
frame
```

```python
frame.sort_values(by='b')
```

```python
frame.sort_values(by=['a','b'])     #정렬 1순위: a, 2순위: b (a정렬에서 같은 것이 있을 때, b의 순위대로 정렬)
```

```python
obj = pd.Series([7,-5,7,4,2,0,4])
obj.rank()
```

```python
obj.rank(method='first')      #동률일 때, 먼저 나타나는 순서로 정렬
```

```python
obj.rank(ascending=False, method='max')       #동률일 때, 가장 큰 순서로 정렬
```

```python
frame = pd.DataFrame({'b':[4.3,7,-3,2], 'a':[0,1,0,1], 'c':[-2,5,8,-2.5]})
frame
```

```python
frame.rank(axis='columns')
```

**5.3.2 유일값, 값 세기, 멤버십**

```python
obj = pd.Series(['c','a','d','a','a','b','b','c','c'])
uniques = obj.unique()
uniques
```

```python
obj.value_counts()
```

```python
pd.value_counts(obj.values, sort=False)
```

```python
obj
```

```python
mask = obj.isin(['b','c'])
```

```python
obj[mask]
```

```python
to_match = pd.Series(['c','a','b','b','c','a'])
unique_vals = pd.Series(['c','b','a'])

pd.Index(unique_vals).get_indexer(to_match)     #to_match와 공통되는 unique_vals 값 인데스값 찾기
```

```python
data = pd.DataFrame({'Qu1':[1,3,4,3,4], 'Qu2':[2,3,1,2,3], 'Qu3':[1,5,2,4,4]})
data
```

```python
result = data.apply(pd.value_counts).fillna(0)
result       #결괏값의 로우 라벨은 전체 컬럼의 유일한 값들을 담고 있다.
             #각 값은 각 컬럼에서 해당 값이 몇 번 출현했는지 나타낸다.
```
