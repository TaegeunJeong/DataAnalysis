**7.2.2 함수나 매핑을 이용해서 데이터 변형하기**

````python
data = pd.DataFrame({'food':['bacon','pulled pork','bacon','Pastrami','corned beef','Bacon','pastrami','honey ham','nova lox'],
                     'ounces':[4,3,12,6,7.5,8,3,5,6]})
data
````

````python
meat_to_animal = {
    'bacon': 'pig',
    'pulled pork': 'pig',
    'pastrami': 'cow',
    'corned beef': 'cow',
    'honey ham': 'pig',
    'nova lox': 'salmon'
}
````

````python
lowercased = data['food'].str.lower()
lowercased
````

````python
data['animal'] = lowercased.map(meat_to_animal)
data
````

````python
data['food'].map(lambda x: meat_to_animal[x.lower()])
data

#map 메서드를 사용하면 데이터의 요소별 변환 및 데이터를 다듬는 작업을 편리하게 수행할 수 있다.
````

````python
mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('https://raw.githubusercontent.com/wesm/pydata-book/2nd-edition/datasets/movielens/movies.dat',sep='::',header=None, names=mnames)
````

````python
movies[:10]
````

````python
all_genres = []
for x in movies.genres:
  all_genres.extend(x.split('|'))

genres = pd.unique(all_genres)
genres
````

````python
zero_matrix = np.zeros((len(movies), len(genres)))
dummies = pd.DataFrame(zero_matrix, columns=genres)
````

````python
gen = movies.genres[0]
gen.split('|')
````

````python
dummies.columns.get_indexer(gen.split('|'))     #아래의 text참고하기. (get_indexer: index값 찾기)
````

to_match와 공통되는 unique_vals 값 인덱스값 찾기
```
to_match = pd.Series(['c','a','b','b','c','a'])
unique_vals = pd.Series(['c','b','a'])

pd.Index(unique_vals).get_indexer(to_match)
```

````python
for i, gen in enumerate(movies.genres):
  indices = dummies.columns.get_indexer(gen.split('|'))
  dummies.iloc[i, indices] = 1
````

````python
movies_windic = movies.join(dummies.add_prefix('Genre_'))
movies_windic.iloc[0]
````

**7.3 문자열 다루기**

````python
val = 'a,b, guido'
val.split(',')
````

````python
pieces = [x.strip() for x in val.split(',')]      #strip: 좌,우 공백문자 없애기
pieces
````

````python
first, second, third = pieces
first+'::'+second+'::'+third
````

````python
'::'.join(pieces)
````

````python
'guido' in val
````

````python
val.index(',')      #찾지 못하면 예외를 발생시킴.
````

````python
val.find(':')       #찾지 못하면 -1을 반환함. (예외 발생 X)
````

````python
val.count(',')
````

````python
val.replace(',','::')
````

````python
val.replace(',','')     #찾아낸 패턴을 다른 문자로 치환함.
                        #비어있는 문자열로 문자열을 대체하여, 패턴을 삭제하는 방법으로 자주 사용됨.
````
